```python
import pandas as pd
import matplotlib.pyplot as plt

# Ensures plots show inside the notebook
%matplotlib inline
df = pd.read_csv("Shopping Trend Assessment Dataset (1).csv")
df.head()
```

```python
age_avg = df.groupby("Age")["Purchase Amount (USD)"].mean()

plt.figure()
plt.plot(age_avg.index, age_avg.values)
plt.xlabel("Age")
plt.ylabel("Average Purchase Amount (USD)")
plt.title("Average Purchase Amount by Age")
plt.show()
```

```python
plt.figure()
plt.scatter(df["Age"], df["Purchase Amount (USD)"])
plt.xlabel("Age")
plt.ylabel("Purchase Amount (USD)")
plt.title("Age vs Purchase Amount")
plt.show()
```

```python
avg_by_category = df.groupby("Category")["Purchase Amount (USD)"].mean()

plt.figure()
avg_by_category.plot(kind="bar")
plt.xlabel("Category")
plt.ylabel("Average Purchase Amount (USD)")
plt.title("Average Purchase Amount by Category")
plt.show()
```
